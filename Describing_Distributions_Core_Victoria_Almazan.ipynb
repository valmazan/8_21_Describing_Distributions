### Loading Data

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy import stats

sns.set_context('talk')
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRSZfNHUJVWx0IKeUo9sQPrE5z-Kwsgq-jjCJEcd_0je91Br6X1_bDaGVjLuChfLGh0CU8uWVvbIHRT/pub?gid=1606036051&single=true&output=csv')
df.info()
df.head()
```

## VitD_levels

```python
col='VitD_levels'
```

```python
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std


fix,ax = plt.subplots(figsize=(10,6))
sns.histplot(data= df, x='VitD_levels', kde= True)
ax.axvline(mean,color = 'red',label='Mean')
ax.axvline(median,color = 'green',ls='--',label='Median')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder=0)

ax.legend();
```

```python
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt
```

 - Is it Discrete or Continuous?

 Continuous


 - Does it have a skew? If so, which direction (+/-)

 For VitD_levels there is now skew and is showing a normal curve. 
    

- What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

 Mesokurtic
